```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
independ=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independ
```

```python
depend=dataset[['charges']]
depend
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'splitter':['best','random'],'max_features':['sqrt','log2']}
```

```python
grid=GridSearchCV(DecisionTreeRegressor(),param,refit=True,verbose=3,n_jobs=-1)
grid.fit(independ,depend)
```

```python
res=grid.cv_results_
print("The R_Score value for bet parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(res)
table
```

```python
Age=input("Age:")
BMI=input("BMI:")
children=input("Children:")
sex=input("Male 0 or1:")
smoke=input("Smoke yes 0 or 1:")
```

```python
Future_Prediction=grid.predict([[Age,BMI,children,sex,smoke]])
print("Future_Prediction={}".format(Future_Prediction))
```
