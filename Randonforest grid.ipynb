```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
independ=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independ
```

```python
depend=dataset[['charges']]
depend
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param={'criterion':('squared_error', 'absolute_error', 'friedman_mse', 'poisson'),'max_features':('auto','sqrt','log2'),'n_estimators':(10,100)}
```

```python
grid=GridSearchCV(RandomForestRegressor(),param,refit=True,verbose=3,n_jobs=-1)
grid.fit(independ,depend)
```

```python
result=grid.cv_results_
print("The R score value the for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
Age=input("Age:")
BMI=input("BMI:")
children=input("Children:")
sex_male=input("Sex male 0 or 1:")
smoke_yes=input("Smoke yes 0 or 1:")
```

```python
Future_Prediction=grid.predict([[Age,BMI,children,sex_male,smoke_yes]])
print("Future_Prediction={}".format(Future_Prediction))
```
